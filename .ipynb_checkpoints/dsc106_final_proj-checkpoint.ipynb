### Bar Graph - Genre

```python
temp = movies[["Netflix", "Hulu", "Prime Video", "Disney+", "Genres"]]

# netflix 
bar_n = pd.DataFrame((temp.loc[temp["Netflix"] == 1]).explode("Genres").groupby("Genres").count()[1:]["Netflix"])
bar_n["Streaming_Service"] = ["Netflix"]*len(bar_n)
bar_n = bar_n.rename(columns={"Netflix": "count"})
bar_n["count"] = bar_n["count"]/len(netflix_movies)

# hulu
bar_h = pd.DataFrame((temp.loc[temp["Hulu"] == 1]).explode("Genres").groupby("Genres").count()[1:]["Hulu"])
bar_h["Streaming_Service"] = ["Hulu"]*len(bar_h)
bar_h = bar_h.rename(columns={"Hulu": "count"})
bar_h["count"] = bar_h["count"]/len(hulu_movies)

# prime
bar_p = pd.DataFrame((temp.loc[temp["Prime Video"] == 1]).explode("Genres").groupby("Genres").count()[1:]["Prime Video"])
bar_p["Streaming_Service"] = ["Prime Video"]*len(bar_p)
bar_p = bar_p.rename(columns={"Prime Video": "count"})
bar_p["count"] = bar_p["count"]/len(prime_movies)

# disney
bar_d = pd.DataFrame((temp.loc[temp["Disney+"] == 1]).explode("Genres").groupby("Genres").count()[1:]["Disney+"])
bar_d["Streaming_Service"] = ["Disney+"]*len(bar_d)
bar_d = bar_d.rename(columns={"Disney+": "count"})
bar_d["count"] = bar_d["count"]/len(disney_movies)


bar = pd.concat([bar_n, bar_h, bar_p, bar_d], axis = 0)
bar["Genres"] = bar.index

color_streaming = alt.Color('Streaming_Service:O', scale=alt.Scale(scheme='tableau10'))

alt.Chart(bar).mark_bar().encode(
    x='Streaming_Service:O',
    y='count:Q',
    color=color_streaming,
    column='Genres:N').properties( title = "Proportion of Movies in Each Genre")

```

### Language

```python
temp = movies[["Netflix", "Hulu", "Prime Video", "Disney+", "Language"]]
top_langs = (movies["Language"].explode().value_counts())[:8].index.tolist()
```

```python
netflix_lang = temp.loc[temp["Netflix"] == 1]
hulu_lang = temp.loc[temp["Hulu"] == 1]
prime_lang = temp.loc[temp["Prime Video"] == 1]
disney_lang = temp.loc[temp["Disney+"] == 1]

n_top = pd.DataFrame(netflix_lang["Language"].explode().value_counts().ix[top_langs]/len(netflix_movies))
h_top = pd.DataFrame(hulu_lang["Language"].explode().value_counts().ix[top_langs]/len(hulu_movies))
p_top = pd.DataFrame(prime_lang["Language"].explode().value_counts().ix[top_langs]/len(prime_movies))
d_top = pd.DataFrame(disney_lang["Language"].explode().value_counts().ix[top_langs]/len(disney_movies))

n_top["Streaming_Service"] = ["Netflix"]*len(n_top)
h_top["Streaming_Service"] = ["Hulu"]*len(h_top)
p_top["Streaming_Service"] = ["Prime Video"]*len(p_top)
d_top["Streaming_Service"] = ["Disney+"]*len(d_top)

languages = pd.concat([n_top, h_top, p_top, d_top], axis = 0)
languages["language"] = languages.index
```

```python
color_streaming = alt.Color('Streaming_Service:O', scale=alt.Scale(scheme='tableau10'))

alt.Chart(languages).mark_bar().encode(
    x='Streaming_Service:O',
    y='Language:Q',
    color=color_streaming,
    column='language:N').properties( title = "Language Perportion in Each Service (Top 8 most common Languages)")

```

## Choropleth - Service with Most movies offered in Each Country

```python
per_country = movies[["Netflix", "Hulu", "Prime Video"
                      , "Disney+", "Country"]].explode("Country").groupby("Country").sum()
per_country["name"] = per_country.index
per_country = pd.merge(countries, per_country, on = "name")
```

```python
temp = per_country[["id", "alpha2", "alpha3", "name", "Netflix", "Hulu", "Prime Video", "Disney+"]]
```

```python
temp["biggest_service"] = per_country[["Netflix", "Hulu", "Prime Video", "Disney+"]].idxmax(axis=1)
```

```python
# base
base_map = alt.Chart(countries_topo).mark_geoshape(fill='lightgray',
    stroke='darkgray').project(type = 'equirectangular').properties(
    width = 500, height = 300)

# Customize Color
color_biggest_service = alt.Color('biggest_service:N'
                                  , scale = alt.Scale(domain = ['null', 'Disney+', 'Netflix', 'Prime Video']
                                                      , range = ['lightgray', '#b5cf6b', '#9ecae1','#9e9ac8']))

# Choro --> Country is color of biggest service 
choro_p = alt.Chart(countries_topo).mark_geoshape().encode(
    color = color_biggest_service).transform_lookup(
    lookup ='id',from_=alt.LookupData(
        temp,'id', ['biggest_service'])).project(
    type = 'equirectangular').properties(width = 500, height = 300, title = "Service with Most Movies Offered in each Country")



base_map + choro_p
```

## Scatter

```python
temp = movies[["Year", "Age", "IMDb", "Rotten Tomatoes", "Netflix", "Hulu", "Prime Video", "Disney+", "Runtime"]].dropna()
```

```python
# Rating column that combines IMDb and Rotten Tomatoes Score
rating_col = (temp["IMDb"] + temp["IMDb"])/2

# Categorize into 3 groups to make variable ordinal rather than quantitative
bad = rating_col[(rating_col< 0.55)].values
ok = rating_col[(rating_col >= 0.55) & (rating_col < 0.75)].values
excellent = rating_col[(rating_col>= 0.75)].values

rating_col = rating_col.replace(bad, "bad ( < 0.55)")
rating_col = rating_col.replace(ok, "good (0.55 ~ 0.75)")
rating_col = rating_col.replace(excellent, "excellent ( > 0.75)")

temp["rating"] = rating_col
```

```python
# properties for all charts
selection = alt.selection_multi(toggle=False, empty='none', fields=['Age'])
color_age = alt.condition(selection,alt.Color('Age:N', legend=None),alt.value('lightgray'))
shape_age = alt.Shape('Age:N', legend=None)
opacity_rating = alt.Opacity('rating:O')
x=alt.X('Year:O', title='Year of Film Release')

# Netflix
temp_n = temp.loc[temp["Netflix"] == 1]

scatter_n = alt.Chart(temp_n).mark_point().encode(
    x=x,y='Runtime:Q',color= color_age, shape = shape_age
    , opacity=opacity_rating).properties(width=300,height=300, title = "Netflix").interactive()
legend_n = alt.Chart(temp_n).mark_point(size=100).encode(
    y=alt.Y('Age:O', title='Age Restrictions',axis=alt.Axis(orient='right')),color='Age:O',shape = "Age:O"
    ,opacity = "rating:O").add_selection(selection)

netflix = scatter_n | legend_n
netflix

# Hulu
temp_h = temp.loc[temp["Hulu"] == 1]

scatter_h = alt.Chart(temp_h).mark_point().encode(
    x=x,y='Runtime:Q',color= color_age, shape = shape_age
    , opacity=opacity_rating).properties(width=300,height=300, title = "Hulu").interactive()
legend_h = alt.Chart(temp_h).mark_point(size=100).encode(
    y=alt.Y('Age:O', title='Age Restrictions', axis=alt.Axis(orient='right')),color='Age:O',shape = "Age:O"
    ,opacity = "rating:O").add_selection(selection)

hulu = scatter_h | legend_h
hulu

# Prime Video
temp_p = temp.loc[temp["Prime Video"] == 1]

scatter_p = alt.Chart(temp_p).mark_point().encode(
    x=x,y='Runtime:Q',color= color_age, shape = shape_age
    , opacity=opacity_rating).properties(width=300,height=300, title = "Prime Video").interactive()
legend_p = alt.Chart(temp_p).mark_point(size=100).encode(
    y=alt.Y('Age:O', title='Age Restrictions', axis=alt.Axis(orient='right')),color='Age:O',shape = "Age:O"
    ,opacity = "rating:O").add_selection(selection)

prime = scatter_p | legend_p
prime

# Disney+
temp_d = temp.loc[temp["Disney+"] == 1]

scatter_d = alt.Chart(temp_d).mark_point().encode(
    x=x,y='Runtime:Q',color= color_age, shape = shape_age
    , opacity=opacity_rating).properties(width=300,height=300, title = "Disney+").interactive()
legend_d = alt.Chart(temp_d).mark_point(size=100).encode(
    y=alt.Y('Age:O', title='Age Restrictions', axis=alt.Axis(orient='right')),color='Age:O',shape = "Age:O"
    ,opacity = "rating:O").add_selection(selection)

disney = scatter_d | legend_d
disney


alt.hconcat(alt.vconcat(netflix, hulu), alt.vconcat(prime, disney))
```
